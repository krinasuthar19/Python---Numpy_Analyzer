```python
import numpy as np
import sys

class DataAnalytics:
    def __init__(self, array=None):
        self._array = np.array(array) if array is not None else None

    @classmethod
    def from_list(cls, lst, shape=None):
        arr = np.array(lst)
        if shape is not None:
            arr = arr.reshape(shape)
        return cls(arr)

    @staticmethod
    def _to_numbers(tokens):
        return [float(x) for x in tokens]

    def _ensure(self):
        if self._array is None:
            raise ValueError("No array set")

    def set_array(self, array):
        self._array = np.array(array)

    def get_array(self):
        return self._array

    def create_1d(self, elements):
        self._array = np.array(elements)

    def create_2d(self, rows, cols, elements):
        arr = np.array(elements)
        arr = arr.reshape((rows, cols))
        self._array = arr

    def create_3d(self, d1, d2, d3, elements):
        arr = np.array(elements)
        arr = arr.reshape((d1, d2, d3))
        self._array = arr

    def index_element(self, idx):
        self._ensure()
        return self._array[tuple(idx)]

    def slice_array(self, slices):
        self._ensure()
        return self._array[tuple(slices)]

    def concatenate(self, other, axis=0):
        self._ensure()
        other = np.array(other)
        return np.concatenate((self._array, other), axis=axis)

    def split_array(self, indices_or_sections, axis=0):
        self._ensure()
        if isinstance(indices_or_sections, int):
            return np.split(self._array, indices_or_sections, axis=axis)
        return np.split(self._array, indices_or_sections, axis=axis)

    def add(self, other):
        self._ensure()
        return np.add(self._array, np.array(other))

    def subtract(self, other):
        self._ensure()
        return np.subtract(self._array, np.array(other))

    def multiply(self, other):
        self._ensure()
        return np.multiply(self._array, np.array(other))

    def divide(self, other):
        self._ensure()
        return np.divide(self._array, np.array(other))

    def dot(self, other):
        self._ensure()
        return np.dot(self._array, np.array(other))

    def matmul(self, other):
        self._ensure()
        return self._array @ np.array(other)

    def search(self, value):
        self._ensure()
        return np.argwhere(self._array == value)

    def sort(self, axis=None, kind='quicksort'):
        self._ensure()
        return np.sort(self._array, axis=axis, kind=kind)

    def filter_condition(self, operator, threshold):
        self._ensure()
        ops = {
            '>': lambda x: x > threshold,
            '<': lambda x: x < threshold,
            '>=': lambda x: x >= threshold,
            '<=': lambda x: x <= threshold,
            '==': lambda x: x == threshold,
            '!=': lambda x: x != threshold
        }
        if operator not in ops:
            raise ValueError("Invalid operator")
        mask = ops[operator](self._array)
        return self._array[mask]

    def aggregate(self, func, axis=None):
        self._ensure()
        funcs = {
            'sum': np.sum,
            'mean': np.mean,
            'median': np.median,
            'std': np.std,
            'var': np.var,
            'min': np.min,
            'max': np.max
        }
        if func not in funcs:
            raise ValueError("Unsupported aggregate")
        return funcs[func](self._array, axis=axis)

    def percentiles(self, percents, axis=None):
        self._ensure()
        return np.percentile(self._array, percents, axis=axis)

    def correlation_coefficient(self, other):
        self._ensure()
        other = np.array(other)
        a = self._array.flatten()
        b = other.flatten()
        if a.size != b.size:
            raise ValueError("Arrays must have same number of elements for correlation")
        return np.corrcoef(a, b)

def input_numbers(prompt, expected_count=None):
    s = input(prompt).strip()
    tokens = s.split()
    nums = DataAnalytics._to_numbers(tokens)
    if expected_count is not None and len(nums) != expected_count:
        raise ValueError("Incorrect number of elements entered")
    return nums

def parse_range(r):
    parts = r.split(':')
    if len(parts) != 2:
        raise ValueError("Range must be in start:end format")
    return int(parts[0]), int(parts[1])

def main_menu():
    da = DataAnalytics()
    while True:
        print("Welcome to the NumPy Analyzer!")
        print("Choose an option:")
        print("1. Create a Numpy Array")
        print("2. Perform Mathematical Operations")
        print("3. Combine or Split Arrays")
        print("4. Search, Sort, or Filter Arrays")
        print("5. Compute Aggregates and Statistics")
        print("6. Exit")
        choice = input("Enter your choice: ").strip()
        if choice == '1':
            create_menu(da)
        elif choice == '2':
            math_menu(da)
        elif choice == '3':
            combine_split_menu(da)
        elif choice == '4':
            search_sort_filter_menu(da)
        elif choice == '5':
            aggregates_menu(da)
        elif choice == '6':
            print("Exiting.")
            sys.exit(0)
        else:
            print("Invalid choice")

def create_menu(da):
    print("Select the type of array to create:")
    print("1. 1D Array")
    print("2. 2D Array")
    print("3. 3D Array")
    ch = input("Enter your choice: ").strip()
    try:
        if ch == '1':
            count = int(input("Enter number of elements: ").strip())
            nums = input_numbers("Enter elements separated by space: ", count)
            da.create_1d(nums)
            print("Array created successfully:")
            print(da.get_array())
        elif ch == '2':
            rows = int(input("Enter the number of rows: ").strip())
            cols = int(input("Enter the number of columns: ").strip())
            nums = input_numbers(f"Enter {rows*cols} elements for the array separated by space: ", rows*cols)
            da.create_2d(rows, cols, nums)
            print("Array created successfully:")
            print(da.get_array())
            indexing_slicing_submenu(da)
        elif ch == '3':
            d1 = int(input("Enter dimension 1 size: ").strip())
            d2 = int(input("Enter dimension 2 size: ").strip())
            d3 = int(input("Enter dimension 3 size: ").strip())
            nums = input_numbers(f"Enter {d1*d2*d3} elements separated by space: ", d1*d2*d3)
            da.create_3d(d1, d2, d3, nums)
            print("Array created successfully:")
            print(da.get_array())
        else:
            print("Invalid choice")
    except Exception as e:
        print("Error:", e)

def indexing_slicing_submenu(da):
    while True:
        print("Choose an operation: 1. Indexing 2. Slicing 3. Go Back")
        c = input("Enter your choice: ").strip()
        if c == '1':
            try:
                raw = input("Enter indices separated by comma for each axis (e.g. 0,1): ").strip()
                idx = tuple(int(x) for x in raw.split(','))
                print("Element:", da.index_element(idx))
            except Exception as e:
                print("Error:", e)
        elif c == '2':
            try:
                rows = input("Enter the row range (start:end): ").strip()
                cols = input("Enter the column range (start:end): ").strip()
                r0, r1 = parse_range(rows)
                c0, c1 = parse_range(cols)
                print("Sliced Array:")
                print(da.slice_array((slice(r0, r1), slice(c0, c1))))
            except Exception as e:
                print("Error:", e)
        elif c == '3':
            break
        else:
            print("Invalid choice")

def math_menu(da):
    try:
        da._ensure()
    except:
        print("No array exists. Create an array first.")
        return
    print("Choose a mathematical operation:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Dot Product")
    print("6. Matrix Multiplication")
    ch = input("Enter your choice: ").strip()
    try:
        if ch in ['1','2','3','4']:
            arr = da.get_array()
            elems = input_numbers(f"Enter the same-size array elements ({arr.size} elements separated by space): ", arr.size)
            other = np.array(elems).reshape(arr.shape)
            if ch == '1':
                res = da.add(other)
            elif ch == '2':
                res = da.subtract(other)
            elif ch == '3':
                res = da.multiply(other)
            else:
                res = da.divide(other)
            print("Original Array:")
            print(arr)
            print("Second Array:", other)
            print("Result:")
            print(res)
        elif ch == '5':
            other = input_numbers("Enter elements for other array (compatible for dot product): ")
            other_arr = np.array(other)
            res = da.dot(other_arr.reshape(other_arr.shape))
            print("Dot product result:")
            print(res)
        elif ch == '6':
            other = input_numbers("Enter elements for other matrix in row-major order: ")
            other_arr = np.array(other)
            try:
                other_arr = other_arr.reshape((-1, da.get_array().shape[0]))
            except:
                pass
            res = da.matmul(other_arr)
            print("Matrix multiplication result:")
            print(res)
        else:
            print("Invalid choice")
    except Exception as e:
        print("Error:", e)

def combine_split_menu(da):
    try:
        da._ensure()
    except:
        print("No array exists. Create an array first.")
        return
    print("Choose an option: 1. Combine Arrays 2. Split Array")
    c = input("Enter your choice: ").strip()
    try:
        if c == '1':
            arr = da.get_array()
            elems = input_numbers(f"Enter the elements of another array to combine ({arr.size} elements separated by space): ", arr.size)
            other = np.array(elems).reshape(arr.shape)
            print("Original Array:")
            print(arr)
            print("Second Array:")
            print(other)
            combined = da.concatenate(other, axis=0 if arr.ndim>=2 else None)
            print("Combined Array (Vertical Stack):")
            print(combined)
        elif c == '2':
            arr = da.get_array()
            if arr.ndim == 1:
                n = int(input("Enter number of splits: ").strip())
                parts = da.split_array(n)
                for p in parts:
                    print(p)
            else:
                axis = int(input("Enter axis to split along (0 or 1 or ...): ").strip())
                mode = input("Split by (1) number of sections or (2) indices? Enter 1 or 2: ").strip()
                if mode == '1':
                    n = int(input("Enter number of sections: ").strip())
                    parts = da.split_array(n, axis=axis)
                else:
                    raw = input("Enter split indices separated by comma: ").strip()
                    idx = [int(x) for x in raw.split(',')]
                    parts = da.split_array(idx, axis=axis)
                for p in parts:
                    print(p)
        else:
            print("Invalid choice")
    except Exception as e:
        print("Error:", e)

def search_sort_filter_menu(da):
    try:
        da._ensure()
    except:
        print("No array exists. Create an array first.")
        return
    print("Choose an option: 1. Search a value 2. Sort the array 3. Filter values")
    c = input("Enter your choice: ").strip()
    try:
        if c == '1':
            v = float(input("Enter value to search: ").strip())
            res = da.search(v)
            print("Found indices:")
            print(res)
        elif c == '2':
            arr = da.get_array()
            print("Original Array:")
            print(arr)
            axis_input = input("Enter axis to sort along or press enter for flattened sort: ").strip()
            axis = None if axis_input=='' else int(axis_input)
            sorted_arr = da.sort(axis=axis)
            print("Sorted Array:")
            print(sorted_arr)
            print("(Sorting applied along the specified axis.)")
        elif c == '3':
            print("Enter operator from >, <, >=, <=, ==, !=")
            op = input("Operator: ").strip()
            thresh = float(input("Threshold value: ").strip())
            res = da.filter_condition(op, thresh)
            print("Filtered values:")
            print(res)
        else:
            print("Invalid choice")
    except Exception as e:
        print("Error:", e)

def aggregates_menu(da):
    try:
        da._ensure()
    except:
        print("No array exists. Create an array first.")
        return
    print("Choose an aggregate/statistical operation:")
    print("1. Sum")
    print("2. Mean")
    print("3. Median")
    print("4. Standard Deviation")
    print("5. Variance")
    print("6. Min")
    print("7. Max")
    print("8. Percentiles")
    print("9. Correlation Coefficient with another array")
    ch = input("Enter your choice: ").strip()
    try:
        if ch in [str(i) for i in range(1,8)]:
            funcs = {'1':'sum','2':'mean','3':'median','4':'std','5':'var','6':'min','7':'max'}
            func = funcs[ch]
            axis = input("Enter axis or press enter for global: ").strip()
            axis = None if axis=='' else int(axis)
            res = da.aggregate(func, axis=axis)
            print(f"Result of {func}:")
            print(res)
        elif ch == '8':
            praw = input("Enter percentiles separated by space (e.g. 25 50 75): ").strip()
            ps = [float(x) for x in praw.split()]
            axis = input("Enter axis or press enter for global: ").strip()
            axis = None if axis=='' else int(axis)
            res = da.percentiles(ps, axis=axis)
            print("Percentiles:")
            print(res)
        elif ch == '9':
            arr = da.get_array()
            elems = input_numbers(f"Enter {arr.size} elements for the other array separated by space: ", arr.size)
            other = np.array(elems).reshape(arr.shape)
            res = da.correlation_coefficient(other)
            print("Correlation coefficient matrix:")
            print(res)
        else:
            print("Invalid choice")
    except Exception as e:
        print("Error:", e)

if __name__ == "__main__":
    main_menu()
```
